```python
#calculator using function
def add(a,b):
 return(a+b)
def sub(a,b):
 return(a-b)
def multiple(a,b):
 return(a*b)
def division(a,b):
 return(a/b)
c=int(input('''Enter
   1.add
   2.sub
   3.multiple
   4.division
    '''))
num1=int(input('Num¹ '))
num2=int(input('Num² '))
if c==1:
  A=add(num1,num2)
  print(A)
elif c==2:
  B=sub(num1,num2)
  print (B)
elif c==3:
  C=multiple(num1,num2)
  print (C)
elif c==4:
  D=division(num1,num2)
  print (D)
else:
  print('ERROR')
```
