```python

#nested while
p=1
while p<11:
 q=1
 while q<=p:
  print (p,end=" ")
  q+=1
 p+=1
 print (" ")
```

```python

#nested while loop
p=10
while p>1:
 q=10
 while q>=p:
  print (p,end=" ")
  q-=1
 p-=1
 print (" ")
```
